```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load the data
dataset = pd.read_csv('/kaggle/input/salary-dataset-simple-linear-regression/Salary_dataset.csv')
```

```python
#Scatter plot of the raw data
plt.scatter(dataset.YearsExperience,dataset.Salary)
```

```python
x = dataset.YearsExperience
y = dataset.Salary
```

```python
#Normalize the input data for better fitting
x = (x-min(x))/(max(x)-min(x))
```

```python
#Gradient-Descent function for SLR
def Gradient_Descent(x,y,L,epochs,m):
    w_grad = 0
    b_grad = 0
    w = 0
    b = 0
    
    for e in range(epochs):
        for i in range(m):
            curr_w = w
            curr_b = b
            fx = curr_w * x[i] + b
            w_grad += ((1/m)*(fx-y[i])*x[i])
            b_grad += ((1/m)*(fx-y[i]))
            
        w = curr_w - (w_grad * L)        
        b = curr_b - (b_grad * L)
        
    return w,b        
```

```python
#Get the weights and biases
w,b = Gradient_Descent(x,y,0.0000021,500,len(x))
```

```python
#Plot the Regression line along with data
plt.scatter(dataset.YearsExperience,dataset.Salary)
plt.plot(list(range(0,12)),[w * x + b for x in range(0,12)],color = "red")
plt.show()
```
